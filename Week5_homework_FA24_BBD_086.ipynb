```python
#Setup Code(importing and setting up our libraries)
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import zscore
from scipy.stats.mstats import winsorize
```

```python
#Importing data and getting basic information
titanic_df = pd.read_csv('/content/Titanic-Dataset.csv')
titanic_df.info()
```

```python
# Activity 1:  Create a histogram of the 'Fare' column with 30 bins and a different color.
plt.hist(titanic_df['Fare'], bins=30, color='pink', edgecolor='black', alpha=0.7)
plt.xlabel('Fare')
plt.ylabel('Frequency')
plt.title('Histogram of Fare')
plt.grid(True)
plt.show()
```

```python
embarked_cnt = titanic_df.groupby('Embarked').size()
embarked_cnt.plot(kind='bar',  color='purple', edgecolor='black', alpha=0.5)
plt.xlabel('Port Embarked')
plt.ylabel('No. of Passengers')
plt.title('No. of Passengers who embarked at each port')
plt.grid(True)
plt.show()
```

```python
# What does this chart shows. What it help us to understand :
#Answer: The chart helps us ascertain the number of passengers that departed from each port. The most passengers departing from SouthHampton and that is shown by 'S' having the tallest bar.
```

```python
# Activity 3: Create a pie chart showing the proportion of male and female survivors.
survivors_df = titanic_df[titanic_df['Survived'] == 1]
survivor_sex = survivors_df['Sex'].value_counts()
plt.figure(figsize=(6,6)) #not necessary but this can help us control the size of our figure
plt.pie(survivor_sex.values, labels=survivor_sex.index, autopct='%1.1f%%', colors=['pink', 'skyblue'])
#autopct is basically a parameter of how we want to see our percentage i.e how many significant figures we want
plt.title('Proportion of Male and Female Survivors')
plt.show()



# What does this pie chart illustrate?
#Answer: The pie chart shows us that a larger proportion of females survived the ship sinking
```

```python
# What does this pie chart illustrate?

#Answer: The pie chart shows us that a larger proportion of females survived the ship sinking
```

```python
# Activity 4: Create a scatter plot of 'Age' vs. 'SibSp' (number of siblings/spouses aboard).
plt.scatter(titanic_df['Age'], titanic_df['SibSp'], alpha =0.5, color= 'green')
plt.xlabel('Age')
plt.ylabel('SibSp')
plt.title('Scatter Plot of Age vs. SibSp')
plt.show()



# What does scatterplot shows?.
#Answer: The highest frequency of Siblings and Spouses is 0
```

```python
# Activity 5: Create a bar chart to compare the average age of survivors and non-survivors
avg_age_by_survival = titanic_df.groupby('Survived')['Age'].agg('mean')
avg_age_by_survival.plot(kind='bar', color=['maroon', 'gray'])
plt.xlabel('Survival Status (0 = No, 1 = Yes)')
plt.ylabel('Average Age')
plt.title('Average Age of Survivors vs. Non-Survivors')
plt.show()
# Answer: What does this bar chart compares?
#The chart compares the average age of survivors vs non survivors, since the avg age of survivors is lower we can assume the younger people were more likely to survive.
```

```python
# Activity 6:  Create a horizontal bar chart that displays the number of passengers in each passenger class (Pclass).
pclass_count = titanic_df['Pclass'].value_counts().sort_index()
plt.barh(pclass_count.index, pclass_count.values, color=['skyblue', 'lightgreen', 'lightcoral'])
plt.ylabel('Passenger Class')
plt.xlabel('Number of Passengers')
plt.title('Number of Passengers in Each Class')
plt.yticks(pclass_count.index, ['1st Class', '2nd Class', '3rd Class'])  # Customize y-axis labels
plt.show()
# What does this horizontal bar chart displays.
#Answer: It shows the distribtution of passengers in each class with the highest number of passengers being in third class
```
